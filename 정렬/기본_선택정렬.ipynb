# 선택 정렬 (selection sort)
데이터가 무작위로 있을 때, 이 중 가장 작은 데이터를 선택해 맨 앞에 있는 데이터와 바꾼다. 그 다음으로 작은 데이터를 선택하여 앞에서 두번째 데이터와 바꾸는 과정을 반복한다.

```python
array = [7,5,9,0,3,1,6,2,4,8]

for i in range(len(array)):
    min_index = i
    print('i:', i, '====================')
    for j in range(i+1, len(array)):
        print('j:', j)
        if array[min_index] > array[j]:
            print('change min_index to', j)
            min_index = j
    print(min_index)
    array[i], array[min_index] = array[min_index], array[i]
    print(array)  
```
